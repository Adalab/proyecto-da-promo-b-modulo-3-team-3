De hecho esta misma base de datos es de 2023 y ya estarían desactualizadas a fecha actual.

Decidimos hacer una imputación en función del JobRole y JobLevel

```python
df.groupby(["JobRole", "JobLevel"])["DailyRate"].mean()
```

```python
moda_dailyrate = round(df.groupby(["JobRole","JobLevel"])["DailyRate"].apply(lambda x: x.mode()[0]), 2)
```

```python
# Crear una función para reemplazar los valores nulos con la moda correspondiente
def reemplazar_moda(row):
    if pd.isnull(row['DailyRate']):
        return moda_dailyrate.loc[row['JobRole'], row['JobLevel']]
    else:
        return row['DailyRate']

# Aplicar la función fila por fila
df['DailyRate'] = df.apply(reemplazar_moda, axis=1)

df["DailyRate"].isna().sum()
```
