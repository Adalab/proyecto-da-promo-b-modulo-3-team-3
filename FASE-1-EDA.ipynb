Vamos a renombras el nombre de las columnas EmployeeNumber, EmployeeCount, YearsWithCurrManager,YearsWithCurrManager, WorkLifeBalance y TotalWorkingYears para que esten como en el documento original

Cambiar tipo variable DailyRate, HourlyRate y MonthlyIncome

```python
#En el caso de DailyRate no sólo hay que eliminar el simbolo del dolar y cambiar la coma por un punto
df["DailyRate"] = df["DailyRate"].str.replace('$', '', regex=False)
df["DailyRate"] = df["DailyRate"].str.replace(',', '.', regex=False)
```

```python
df["DailyRate"] = df["DailyRate"].astype(float) 
```

```python
#Aqui todos los valores son número menos "Not Available". Entenemodes que es un nulo 
df["HourlyRate"] = df["HourlyRate"].replace('Not Available', np.nan, regex=False)
```

```python
df["HourlyRate"] = df["HourlyRate"].astype(float)
```

```python
df["MonthlyIncome"] = df["MonthlyIncome"].str.replace(',', '.', regex=False)
```

```python
df["MonthlyIncome"] = df["MonthlyIncome"].astype(float)
```
